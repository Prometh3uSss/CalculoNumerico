Introducción a Julia

Julia es un lenguaje de programación de alto rendimiento, diseñado para la computación numérica y científica. Combina la facilidad de uso de lenguajes como Python y R con la velocidad de lenguajes como C y Fortran. Julia es especialmente popular en áreas como la ciencia de datos, la inteligencia artificial y la computación técnica.

Primeros Pasos:

"Hola Mundo"

El primer programa que generalmente se escribe en cualquier lenguaje es "Hola Mundo". En Julia, se hace de la siguiente manera:

println("Hola, Mundo!")

Variables:
Las variables en Julia se definen simplemente asignando un valor a un nombre de variable. No es necesario declarar el tipo de la variable, ya que Julia es un lenguaje de tipado dinámico.

x = 10           # Un número entero
y = 3.14         # Un número de punto flotante
nombre = "Julia" # Una cadena de texto

Tipos de Datos:
Julia tiene varios tipos de datos, entre los más comunes están:

Int: Números enteros.
Float64: Números de punto flotante.
String: Cadenas de texto.
Bool: Valores booleanos (true o false).

Ejemplo de uso de diferentes tipos de datos:

a = 5          # Int
b = 2.5        # Float64
c = "Hola"     # String
d = true       # Bool

Operaciones Básicas:

Julia permite realizar operaciones matemáticas básicas:

a = 10
b = 5

suma = a + b             # Suma
resta = a - b            # Resta
multiplicacion = a * b   # Multiplicación
division = a / b         # División

Estructuras de Control:

Condicionales:
Las estructuras condicionales se utilizan para ejecutar diferentes bloques de código según ciertas condiciones.

x = 10

if x > 0
    println("x es positivo")
elseif x < 0
    println("x es negativo")
else
    println("x es cero")
end

Bucles:
Los bucles permiten repetir un bloque de código varias veces.

Bucle for:

for i in 1:5
    println("Número: $i")
end

Bucle while:

contador = 1
while contador <= 5
    println("Contador: $contador")
    contador += 1
end

Funciones:
Las funciones son bloques de código que realizan una tarea específica y pueden ser reutilizadas.

function suma(a, b)
    return a + b
end

resultado = suma(3, 4)  # Llama a la función
println("La suma es: $resultado")

Arreglos (Arrays):
Los arreglos son colecciones de elementos del mismo tipo.

arreglo = [1, 2, 3, 4, 5]

# Acceder a elementos
println(arreglo[1])  # Imprime 1

# Modificar elementos
arreglo[2] = 10
println(arreglo)     # Imprime [1, 10, 3, 4, 5]

Diccionarios:
Los diccionarios son colecciones de pares clave-valor.

diccionario = Dict("nombre" => "Julia", "edad" => 5)

# Acceder a valores
println(diccionario["nombre"])  # Imprime "Julia"

# Modificar valores
diccionario["edad"] = 6
println(diccionario)  # Imprime Dict("nombre" => "Julia", "edad" => 6)

Módulos:
Los módulos permiten organizar el código en diferentes archivos y espacios de nombres.

module MiModulo
    export saludar

    function saludar(nombre)
        println("Hola, $nombre!")
    end
end

using .MiModulo
saludar("Mundo")  # Imprime "Hola, Mundo!"
