```python
1. Estructura Básica de un Programa
Un programa en Go comienza con el paquete principal (package main) y la función principal (func main()), que es el punto de entrada del programa.

package main

import "fmt"

func main() {
    fmt.Println("¡Hola, mundo!")
}

2. Paquetes
Paquete: Cada archivo Go pertenece a un paquete. El paquete main es especial porque indica que el programa es ejecutable.
Importar: Para usar funciones de otros paquetes, se utiliza la palabra clave import.

3. Funciones
Las funciones se definen con la palabra clave func, seguida del nombre de la función y los parámetros entre paréntesis.

func suma(a int, b int) int {
    return a + b
}

4. Variables
Las variables se declaran usando var o con la sintaxis corta :=.

var x int = 10 // Declaración tradicional
y := 20       // Declaración corta

5. Tipos de Datos
Go tiene varios tipos de datos básicos, como:

int: Números enteros
float64: Números de punto flotante
string: Cadenas de texto
bool: Valores booleanos (true/false)

6. Estructuras de Control
Condicionales
Se utilizan if, else if y else para las condiciones.

if x > y {
    fmt.Println("x es mayor que y")
} else {
    fmt.Println("x no es mayor que y")
}

Bucles
El bucle for es el único bucle en Go.

for i := 0; i < 5; i++ {
    fmt.Println(i)
}

7. Estructuras
Puedes definir tus propias estructuras con struct.

type Persona struct {
    Nombre string
    Edad   int
}

8. Interfaces
Las interfaces permiten definir métodos que deben ser implementados por otros tipos.

type Animal interface {
    HacerSonido() string
}

9. Manejo de Errores
Go utiliza un enfoque explícito para el manejo de errores, devolviendo un valor de error.


func dividir(a, b int) (int, error) {
    if b == 0 {
        return 0, fmt.Errorf("no se puede dividir por cero")
    }
    return a / b, nil
}
```
